```python
# 67) Write a query to get top 3 directors based on India gross collections in Bollowood industry?

import os
import pandas as pd
from google.colab import files

# ‚úÖ Step 1: Upload Files Manually
print("üìÇ Upload the following CSV files: boxoffice.csv, director.csv, language.csv")
uploaded = files.upload()

# ‚úÖ Step 2: Get the Correct Filenames from Uploaded Files
file_list = os.listdir()
boxoffice_file = [f for f in file_list if "boxoffice" in f and f.endswith(".csv")][0]
director_file = [f for f in file_list if "director" in f and f.endswith(".csv")][0]
language_file = [f for f in file_list if "language" in f and f.endswith(".csv")][0]

print(f"\n‚úÖ Using files:\n- {boxoffice_file}\n- {director_file}\n- {language_file}\n")

# ‚úÖ Step 3: Load CSV Files
boxoffice_df = pd.read_csv(boxoffice_file)
director_df = pd.read_csv(director_file)
language_df = pd.read_csv(language_file)

# ‚úÖ Step 4: Standardize Column Names (Remove Spaces & Lowercase)
boxoffice_df.columns = boxoffice_df.columns.str.strip().str.lower()
director_df.columns = director_df.columns.str.strip().str.lower()
language_df.columns = language_df.columns.str.strip().str.lower()

# ‚úÖ Step 5: Check Column Names
print("üìå Box Office Columns:", boxoffice_df.columns)
print("üìå Director Columns:", director_df.columns)
print("üìå Language Columns:", language_df.columns)

# ‚úÖ Step 6: Merge Data Using Correct Column Names
df = pd.merge(boxoffice_df, director_df, on='directorid', how='left')  # Merge Box Office & Director Data
df = pd.merge(df, language_df, on='languageid', how='left')  # Merge with Language Data

# ‚úÖ Step 7: Filter Bollywood Movies (Industry = Bollywood & Language = Hindi)
df_bollywood = df[(df['industry'].str.contains("Bollywood", case=False, na=False)) &
                  (df['language'].str.contains("Hindi", case=False, na=False))]

# ‚úÖ Step 8: Group by Director & Sum India Gross Collection
top_directors = df_bollywood.groupby('director')['india_gross_collection'].sum().reset_index()

# ‚úÖ Step 9: Get Top 3 Directors Based on India Gross Collection
top_directors = top_directors.sort_values(by='india_gross_collection', ascending=False).head(3)

# ‚úÖ Step 10: Display Results
print("\nüèÜ Top 3 Bollywood Directors Based on India Gross Collections:")
print(top_directors)
```

```python
#74write a query to get genere wise budget
import os
import pandas as pd
from google.colab import files

# ‚úÖ Step 1: Upload Files Manually
print("üìÇ Upload the following CSV files: boxoffice.csv, genre.csv")
uploaded = files.upload()

# ‚úÖ Step 2: Get Correct Filenames from Uploaded Files
file_list = os.listdir()
boxoffice_file = [f for f in file_list if "boxoffice" in f and f.endswith(".csv")][0]
genre_file = [f for f in file_list if "genre" in f and f.endswith(".csv")][0]

print(f"\n‚úÖ Using files:\n- {boxoffice_file}\n- {genre_file}\n")

# ‚úÖ Step 3: Load CSV Files
boxoffice_df = pd.read_csv(boxoffice_file)
genre_df = pd.read_csv(genre_file)

# ‚úÖ Step 4: Standardize Column Names (Remove Spaces & Convert to Lowercase)
boxoffice_df.columns = boxoffice_df.columns.str.strip().str.lower()
genre_df.columns = genre_df.columns.str.strip().str.lower()

# ‚úÖ Step 5: Display Column Names
print("\nüìå Box Office Columns:", list(boxoffice_df.columns))
print("üìå Genre Columns:", list(genre_df.columns))

# ‚úÖ Step 6: Check if 'genre' Column Exists in Box Office Data
if 'genre' in boxoffice_df.columns:
    print("\n‚úÖ Genre column found in 'boxoffice.csv'! No need to merge with 'genre.csv'.")
    use_boxoffice_directly = True
else:
    print("\n‚ùå Genre column NOT found in 'boxoffice.csv'! Checking 'genre.csv' for genreid.")
    use_boxoffice_directly = False

# ‚úÖ Step 7: Check for a Common Merge Key
common_columns = set(boxoffice_df.columns).intersection(set(genre_df.columns))

if 'genreid' in common_columns:
    merge_key = 'genreid'
    print("\n‚úÖ Merging using 'genreid' column.")
elif 'filmid' in common_columns:
    merge_key = 'filmid'
    print("\n‚úÖ Merging using 'filmid' column.")
else:
    print("\n‚ùå ERROR: No common key (genreid/filmid) found for merging! Proceeding with boxoffice.csv only.")
    merge_key = None
    use_boxoffice_directly = True  # Force use of boxoffice.csv directly

# ‚úÖ Step 8: Merge Data (if Necessary)
if not use_boxoffice_directly and merge_key:
    df = pd.merge(boxoffice_df, genre_df, on=merge_key, how='left')
else:
    df = boxoffice_df.copy()  # Use boxoffice.csv directly if merge fails

# ‚úÖ Step 9: Ensure Budget Column Exists
if 'budget' not in df.columns:
    print("\n‚ùå ERROR: 'budget' column not found in dataset! Check column names in 'boxoffice.csv'.")
else:
    # ‚úÖ Step 10: Convert Budget Column to Numeric
    df['budget'] = pd.to_numeric(df['budget'], errors='coerce')

    # ‚úÖ Step 11: Group by Genre and Sum the Budgets
    if 'genre' in df.columns:
        genre_budget_df = df.groupby('genre')['budget'].sum().reset_index()

        # ‚úÖ Step 12: Display Results
        print("\nüí∞ Total Budget per Genre:")
        print(genre_budget_df)

        # ‚úÖ Step 13: Save Output to CSV
        genre_budget_df.to_csv("genre_budget_output.csv", index=False)
        print("\n‚úÖ Output saved as 'genre_budget_output.csv'. Download it from the Files section.")
    else:
        print("\n‚ùå ERROR: 'genre' column still missing after processing. Please check the data files!")


```

```python
#93Write a stored procdure to get industry Budget and Worldwide total collections? If we call any industry it will display industry, budget and wordwide total?
import pandas as pd
from google.colab import files

# üìå Step 1: Upload CSV Files Manually
print("üìÇ Please upload 'boxoffice.csv'")
uploaded = files.upload()

# üìå Step 2: Load Dataset
boxoffice_file = next(iter(uploaded))  # Get the uploaded filename
boxoffice_df = pd.read_csv(boxoffice_file)

# üìå Step 3: Check Columns
print("\n‚úÖ Box Office Dataset Columns:", boxoffice_df.columns.tolist())
print("\nüìä Sample Data:")
print(boxoffice_df.head())

# üìå Step 4: Define a Function (Simulating Stored Procedure)
def get_industry_budget_collections(industry_name=None):
    """
    This function retrieves the total budget and worldwide collection for a specific industry.
    If no industry is given, it returns data for all industries.
    """
    # Group by Industry
    industry_summary = boxoffice_df.groupby('Industry').agg(
        Total_Budget=('Budget', 'sum'),
        Total_Worldwide_Collection=('Worldwide_Collection_in_Crores', 'sum')
    ).reset_index()

    # Display for a specific industry if provided
    if industry_name:
        industry_summary = industry_summary[industry_summary['Industry'].str.lower() == industry_name.lower()]

    return industry_summary

# üìå Step 5: Get Industry-Wise Summary (Full Data)
print("\nüèÜ Industry-Wise Budget and Worldwide Collections:")
print(get_industry_budget_collections())

# üìå Step 6: Get Data for a Specific Industry (Example: Bollywood)
industry_name = "Bollywood"  # Change this value to test different industries
print(f"\nüé¨ {industry_name} Budget and Worldwide Collections:")
print(get_industry_budget_collections(industry_name))
```

```python
#65 Write a query to get top 10 directors based number of films?
# ‚úÖ Step 1: Import required libraries
import pandas as pd
import io
import os
from google.colab import files

# ‚úÖ Step 2: Upload files manually
print("üì§ Please upload 'boxoffice.csv' and 'director.csv'")
uploaded = files.upload()

# ‚úÖ Step 3: Identify correct filenames
file_list = os.listdir()
boxoffice_filename = next((f for f in file_list if "boxoffice" in f.lower()), None)
director_filename = next((f for f in file_list if "director" in f.lower()), None)

# ‚úÖ Step 4: Check if files are uploaded
if not boxoffice_filename or not director_filename:
    raise FileNotFoundError("‚ùå Missing files! Please upload 'boxoffice.csv' and 'director.csv'.")

# ‚úÖ Step 5: Read the files
boxoffice_df = pd.read_csv(boxoffice_filename)
director_df = pd.read_csv(director_filename)

# ‚úÖ Step 6: Standardize column names
boxoffice_df.columns = boxoffice_df.columns.str.lower()
director_df.columns = director_df.columns.str.lower()

# ‚úÖ Step 7: Merge datasets on 'directorid'
merged_df = pd.merge(boxoffice_df, director_df, on="directorid", how="left")

# ‚úÖ Step 8: Count the number of films per director
director_film_count = merged_df.groupby("director")["filmid"].count().reset_index()
director_film_count.columns = ["director", "film_count"]

# ‚úÖ Step 9: Get the top 10 directors based on film count
top_10_directors = director_film_count.sort_values("film_count", ascending=False).head(10)

# ‚úÖ Step 10: Display and save results
print("\nüé¨ Top 10 Directors Based on Number of Films:")
print(top_10_directors)

# Save results to CSV
top_10_directors.to_csv("top_10_directors.csv", index=False)
print("\n‚úÖ Results saved as 'top_10_directors.csv'")

# ‚úÖ Step 11: Allow file download
files.download("top_10_directors.csv")
```

```python
# 66 Write a query to get top 5 directors based on world wide collections and also industry wise?
# ‚úÖ Step 1: Import required libraries
import pandas as pd
import io
import os
from google.colab import files

# ‚úÖ Step 2: Upload files manually
print("üì§ Please upload 'boxoffice.csv' and 'director.csv'")
uploaded = files.upload()

# ‚úÖ Step 3: Identify correct file names (handles spaces or versioning issues)
file_list = os.listdir()
boxoffice_filename = next((f for f in file_list if "boxoffice" in f.lower()), None)
director_filename = next((f for f in file_list if "director" in f.lower()), None)

# ‚úÖ Step 4: Ensure files are uploaded
if not boxoffice_filename or not director_filename:
    raise FileNotFoundError("‚ùå Missing files! Please upload 'boxoffice.csv' and 'director.csv'.")

# ‚úÖ Step 5: Load datasets
boxoffice_df = pd.read_csv(boxoffice_filename)
director_df = pd.read_csv(director_filename)

# ‚úÖ Step 6: Standardize column names
boxoffice_df.columns = boxoffice_df.columns.str.lower()
director_df.columns = director_df.columns.str.lower()

# ‚úÖ Step 7: Merge datasets on 'directorid'
merged_df = pd.merge(boxoffice_df, director_df, on="directorid", how="left")

# ‚úÖ Step 8: Group by Director and Calculate Total Worldwide Collections
director_collections = merged_df.groupby(["director", "industry"])["worldwide_collection_in_crores"].sum().reset_index()

# ‚úÖ Step 9: Get Top 5 Directors Overall
top_5_directors = director_collections.sort_values("worldwide_collection_in_crores", ascending=False).head(5)

# ‚úÖ Step 10: Get Top 5 Directors Per Industry
top_5_per_industry = director_collections.sort_values(["industry", "worldwide_collection_in_crores"], ascending=[True, False])
top_5_per_industry = top_5_per_industry.groupby("industry").head(5)

# ‚úÖ Step 11: Display Results
print("\nüèÜ Top 5 Directors Based on Worldwide Collections:")
print(top_5_directors)

print("\nüé¨ Top 5 Directors Per Industry Based on Collections:")
print(top_5_per_industry)

# ‚úÖ Step 12: Save and Download Results
top_5_directors.to_csv("top_5_directors.csv", index=False)
top_5_per_industry.to_csv("top_5_directors_per_industry.csv", index=False)
print("\n‚úÖ Results saved as 'top_5_directors.csv' and 'top_5_directors_per_industry.csv'")

# ‚úÖ Step 13: Allow file download
files.download("top_5_directors.csv")
files.download("top_5_directors_per_industry.csv")

```
